```python
import numpy as np
```

```python
X = np.linspace(0,100,250) * 0.05
```

```python
X
```

```python
W = 6
B = 5
y = B + W * X
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y)
```

```python
np.random.seed(13)
w, b = np.random.randn(1), np.random.randn(1)
w, b
```

```python
plt.plot(X,y,c='g')
plt.plot(X,w*X+b,c='r')
```

```python
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
Loss_train = []
Loss_test = []
epochs = []
```

```python
from tqdm import tqdm
EPOCH = 2000
lr = 0.01
for epoch in tqdm(range(EPOCH)):
    y_pred = w * X_train + b
    loss = np.mean(np.square(y_train - y_pred))
    dy_pred = -2 * (y_train - y_pred)
    dw = np.mean(X_train * dy_pred)
    db = np.mean(dy_pred)
    w = w - lr * dw
    b = b - lr * db
    if epoch % 10 == 0:
        tqdm.write(f'Train Loss:{loss}')
    loss_test = np.mean(np.square(y_test - w * X_test - b))
    Loss_train.append(loss)
    Loss_test.append(loss_test)
    epochs.append(epoch)
```

```python
plt.plot(epochs,Loss_train,c='r',label='Train Loss')
plt.plot(epochs,Loss_test,c='g',label='Test Loss')
plt.legend()
```

```python
plt.plot(X_train,y_train,c='g',label='Train')
plt.plot(X_test,y_test,c='r',label='Test')
plt.plot(X,w*X+b,c='b',label='Model')
plt.legend()
```
